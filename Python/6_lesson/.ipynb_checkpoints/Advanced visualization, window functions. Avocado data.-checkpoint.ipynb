```python
import pandas as pd
import datetime as dt
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline

sns.set(
    font_scale=2,
    style='whitegrid',
    rc={'figure.figsize' : (21, 5)},
)
```

```python
df = pd.read_csv('avocado_mean.csv', index_col=0, squeeze=True, parse_dates=['Date'])
```

```python
df_ap_mean = df.rolling(window=3).mean()
```

```python
max_value_ap = round(df_ap_mean.max(), 2)
```

```python
import matplotlib.dates as mdates

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10), sharey='col', sharex=True)

windows = [2, 4, 10, 50]
colors = ['blue', 'coral', 'green', 'red']

for window, ax, color in zip(windows, axes.flatten(), colors): 
    ax.plot(df.rolling(window=window).mean(), label=window, color=color)

for ax in axes.flatten():
    # удаляем рамку
    ax.set_frame_on(False)  
    # устанавливаем major locator – 4 января для каждого года
    ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=1, bymonthday=4))  
    # показывать в формате сокращенного названия месяца и дня (Jan 04)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
    # под major locator - minor locator, т.е. редактируем minor ticks
    ax.xaxis.set_minor_locator(mdates.YearLocator(month=1, day=1))
    # показываем год
    ax.xaxis.set_minor_formatter(mdates.DateFormatter('\n%Y'))
    # делаем сетку графика совсем немного серой и наполовину прозрачной
    ax.grid(True, color='#e2e2e2', alpha=0.5)

for name, ax in zip(['1','2','3','4'], axes.flatten()):
    ax.set(title='График {}'.format(name))
    ax.tick_params(labelbottom=True, which='both')

plt.show()
```

```python
delays = pd.read_csv('delays.csv')
```

```python
delays.delay = delays.delay.apply(lambda x: x.split('-'))
```

```python
delays.delay = delays.delay.apply(lambda x: ''.join(x))
```

```python
delays.delay = pd.to_timedelta(delays.delay)
```

```python
delays.dtypes
```

```python
delays['delay_categorical'] = pd.cut(delays.delay, 3)
```

```python
labels = ['less than 1 day', '1-2 days', '2-3 days', 'more than 3 days']

intervals = pd.to_timedelta(['0d', '1d', '2d', '3d', '10d'])
```

```python
delays['delay_categorical'] = pd.cut(delays.delay, bins = intervals, labels=labels)
```

```python
counter = delays.sort_values(by='delay', ascending=True)
```

```python
import plotly.express as px

px.bar(x=delays.delay_categorical, y=delays.delay.value_counts())
```
